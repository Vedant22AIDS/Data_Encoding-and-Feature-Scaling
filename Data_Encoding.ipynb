```python
#single married, in a relationship
#single > [1, 0, 0]
#married > [0, 1, 0]
#separated > [0, 0, 1]
#nominal ohe >> binary vectors for each category
```

```python
df = pd.DataFrame({'status': ['Single', 'Married', 'Single', 'Relationship', 'Married', 'Single']})
```

```python
df
```

```python
encoder=OneHotEncoder()#object of OneHotEncoder()
encoder
```

```python
encoded = encoder.fit_transform(df[['status']]).toarray()

'''
fit will bring mue and sigma and transform will perform calculation
fit -> single,married,relationship
transform ->[1,0,0],[0,1,0],[0,0,1]
'''
```

```python
encoded
```

```python
encoder.get_feature_names_out()
```

```python
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
#new data
encoder.transform([['Single']]).toarray()
```

```python
final_df = pd.concat([df, encoder_df], axis = 1)
```

```python
final_df
```

```python
final_df.drop(['status','status_Married'],inplace=True,axis=1)
```

Target guided ordinal encoding

```python
#target guided ordinal encoding
#relation with target variable
#A lot of unique categories
#replace with mean and and median
```

```python
df = pd.DataFrame({'time': ['lunch', 'breakfast', 'dinner', 'lunch', 'breakfast', 'dinner', 'lunch', 'breakfast', 'dinner'],
             'total_bill': [120, 120, 40, 150, 160, 130,  40, 150, 160]})
```

```python
df
```

```python
mean_price = df.groupby('time')['total_bill'].mean().to_dict()
```

```python
mean_price
```

```python
df['time_target_encoder']=df['time'].map(mean_price)
```
