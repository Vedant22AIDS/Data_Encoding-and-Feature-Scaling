```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

scaling (optional)>>brings all the feature on the same scale
why scaling? --- To reduce the computation time(faster optimisation) and better interpretation

```python
df = sns.load_dataset('tips')
```

```python
df
```

standardisation >> zsore = x-mu/sigma


```python
mean=np.mean(df['total_bill'])
```

```python
mean
```

```python
std=np.std(df['total_bill'])
```

```python
std
```

```python
std_data=[]
for i in list(df['total_bill']):
  zscore=i-mean/std
  std_data.append(zscore)
```

```python
std_data
```

#scaling is optional because it doesnt change the distribution of data
#before scaling

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(std_data)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
#Another Way of doing fit and transform bu sklearn library
scaler.fit(df[['total_bill']]) #calculate mu and sigma
```

```python
scaler.transform(df[['total_bill']])
```

```python
#scaling multiple columns at one time
scaler.fit_transform(df[['total_bill','tip']])
```

```python
pd.DataFrame(scaler.fit_transform(df[['total_bill', 'tip']]), columns = ['total_bill', 'tip'])
```

```python
#new data or unseen data
scaler.transform([[13, 1]])
```
